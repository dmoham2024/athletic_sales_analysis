### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

products_sold = combined_athletic.groupby(['region','state','city',])['units_sold'].sum().reset_index()
products_sold.rename(columns={'units_sold':'Total_Products_Sold'}, inplace=True)

products_sold_sorted = products_sold.sort_values(by='Total_Products_Sold', ascending=False).head(5)

# Show the top 5 results.
print ('Top 5 regions by Total Product Sold')
products_sold_sorted.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_table = pd.pivot_table(combined_athletic, 
                             values='units_sold',
                             index = ['region', 'state', 'city'], 
                             aggfunc = 'sum').reset_index()     

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_table.rename(columns= {'units_sold': 'Total_Products_Sold'}, inplace=True)
pivot_table_sorted = pivot_table.sort_values(by='Total_Products_Sold', ascending=False)
# Show the top 5 results.
pivot_table_sorted.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"

regional_sales = combined_athletic.groupby(['region','state','city',])['total_sales'].sum().reset_index()
regional_sales.rename(columns={'total_sales':'Total_Sales'}, inplace=True)

regional_sales_sorted = regional_sales.sort_values(by='Total_Sales', ascending=False).head(5)




# Show the top 5 results.
regional_sales_sorted.head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.

womens_athletic_footwear = combined_athletic[combined_athletic['product'] == 'Womens Athletic Footwear']
womens_athletic_footwear['invoice_date'] = pd.to_datetime(womens_athletic_footwear['invoice_date'])
daily_sales = womens_athletic_footwear.groupby('invoice_date')['total_sales'].sum().reset_index()

pivot_table_sales_wf = pd.pivot_table(daily_sales,
                                   values='total_sales',
                                   index='invoice_date',
                                   aggfunc='sum')


# Optional: Rename the "total_sales" column to "Total Sales"

pivot_table_sales_wf.rename(columns={'total_sales': 'Total Sales'}, inplace=True)
pivot_table_sales_wf_sorted = pivot_table_sales_wf.sort_index(ascending=False).head(5)



# Show the table.
pivot_table_sales_wf_sorted.head(10)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
pivot_table_dailybins= pivot_table_sales_wf.resample ('D').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
pivot_table_dailybins_sorted= pivot_table_dailybins.sort_values(by='Total_Sales', ascending=False)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
pivot_table_weeklybins= pivot_table_sales_wf.resample ('W').sum()


# Sort the resampled pivot table in descending order on "Total Sales".
pivot_table_weeklybins_sorted=pivot_table_weeklybins.sort_values(by='Total_Sales', ascending=False)
```
